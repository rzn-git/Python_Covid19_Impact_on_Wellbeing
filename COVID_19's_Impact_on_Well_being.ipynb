**COVID-19's Impact on Well-being**

A Data-Driven Examination of the Correlation Between COVID-19 and Happiness Countrywise

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('Modules are imported.')
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/dataset/covid19_Confirmed_dataset.csv")
df.head()
```

```python
df.shape
```

```python
df1=df.drop(["Lat","Long"],axis=1)
```

```python
df1.head()
```

```python
df2=df1.groupby("Country/Region").sum()
pd.set_option('display.max_rows', None)
df2
```

```python
df2.shape
```

```python
df2.loc["Canada"]
```

```python
df2.loc["Canada"].plot()
df2.loc["Germany"].plot()
df2.loc["Bangladesh"].plot()
plt.legend()
```

```python
countries= list(df2.index)
countries
```

```python
type(countries)
```

```python
max_infection_rates = []
for c in countries :
  max_infection_rates.append(df2.loc[c].diff().max())
```

```python
df2["max_infection_rates"] = max_infection_rates
df2.head()
```

```python
df3= pd.DataFrame(df2["max_infection_rates"])
df3.head()
```

```python
dfhp=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/dataset/worldwide_happiness_report.csv")
dfhp.head()
```

```python
useless_col = ["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
dfhp2=dfhp.drop(useless_col,axis=1)
dfhp2.head()
```

```python
dfhp3=dfhp2.set_index("Country or region")
dfhp3.head()
```

```python
df3.head()
```

```python
df3.shape
```

```python
dfhp3.shape
```

```python
df4= df3.join(dfhp3,how='inner')
df4.head()
```
